Este dataframe está compuesto por las siguientes variables:

·**Year**: 2008\
·**Month**: 1-12\
·**DayofMonth**: 1-31\
·**DayOfWeek**: 1 (lunes) - 7 (domingo)\
·**DepTime**: hora de salida real (local, hhmm)\
·**CRSDepTime**: horario de salida programado (local, hhmm)\
·**ArrTime**: hora de llegada real (local, hhmm)\
·**CRSArrTime**: hora de llegada prevista (local, hhmm)\
·**UniqueCarrier**: código único de transportista\
·**FlightNum**: número de vuelo\
·**TailNum**: número de cola del avión: matrícula de la aeronave, identificador único de la aeronave\
·**ActualElapsedTime**: Tiempo transcurrido real: en minutos\
·**CRSElapsedTime**: tiempo estimado en minutos\
·**AirTime**: Tiempo de vuelo en minutos\
·**ArrDelay**: Retraso en la llegada, en minutos: Un vuelo se cuenta como "a tiempo" si operó menos de 15 minutos después de la hora programada\ 
·**DepDelay**: retraso de salida, en minutos\
·**Origen**: código IATA del aeropuerto de origen\
·**Dest**: código de aeropuerto IATA de destino\
·**Distance**: en millas\
·**TaxiIn**: taxi en tiempo, en minutos\
·**TaxiOut**: tiempo de salida del taxi en minutos\
·**Cancelled**: fue cancelado el vuelo, 0 es no y 1 es sí\
·**CancellationCode**: motivo de la cancelación (A = transportista, B = clima, C = NAS, D = seguridad)\
·**Diverted**: desviado: 1 = si, 0 = no\
·**CarrierDelay**: Retraso del transportista en minutos\
·**WeatherDelay**: Demora climática en minutos\
·**NASDelay**: en minutos,el retraso que está bajo el control del Sistema Nacional del Espacio Aéreo (NAS)\ 
·**SecurityDelay**: Demora de seguridad en minutos\
·**LateAircraftDelay**: en minutos,retraso en la llegada a un aeropuerto debido a la llegada tardía de la misma aeronave a un aeropuerto anterior.
